```python
sales = float(input ("Enter sales value ($): "))
tax_rate = float(input ("Enter tax rate (%): "))

COGS= sales * .40
gross_profit= sales - COGS
tax = gross_profit * (tax_rate / 100)
net_profit = gross_profit - tax

print(f"Sales: ${sales}")
print(f"Cost of Goods Sold: ${COGS}")
print(f"Gross Profit: ${gross_profit}")
print(f"Tax: ${tax}")
print(f"Net Profit: ${net_profit}")
```

```python
# Get user inputs
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

# Perform calculation
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero"
else:
    result = "Invalid operator"

# Print result
print("Result:", result)
```

```python
base_fee = 5
hourly_rate = 2.5

print("Hours  Fee")

for hours in range(1, 9):
    fee = base_fee + (hourly_rate * hours)

    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0

    print(hours, "   ", fee)
```

```python
import math

print("Ads   Profit")

for ads in range(0, 201, 25):
    extra_attendees = 2 * math.sqrt(ads)
    total_attendees = 20 + extra_attendees

    revenue = total_attendees * 10
    total_cost = 200 + ads

    profit = revenue - total_cost

    print(f"{ads:3}  {profit:.1f}")
```

```python
def checkEmailAddress(email):
    # Ensure input is a string
    assert isinstance(email, str), "Input must be a string"

    try:
        # Check exactly one "@"
        if email.count("@") != 1:
            return False

        username, domain = email.split("@")

        # Check at least one "." in domain
        if "." not in domain:
            return False

        domain_name, top_domain = domain.rsplit(".", 1)

        # Ensure no part is empty
        if not username or not domain_name or not top_domain:
            return False

        return True

    except:
        return False


# Main program
email = input("Enter email address: ")
print(checkEmailAddress(email))
```

```python
def checkEmailAddress(email):
    # Keep assertion for type checking
    assert isinstance(email, str), "Input must be a string"

    # Raise errors based on requirements
    if email == "":
        raise ValueError("Input cannot be empty")

    if email.count("@") != 1:
        raise ValueError("Email must contain exactly one '@' symbol")

    username, domain = email.split("@")

    if "." not in domain:
        raise ValueError("Domain must contain at least one '.'")

    domain_name, top_domain = domain.rsplit(".", 1)

    if not username or not domain_name or not top_domain:
        raise ValueError("Username, domain, or top-domain cannot be empty")

    return True


# Main program with try-except
try:
    email = input("Enter email address: ")
    result = checkEmailAddress(email)
    print(result)

except AssertionError as ae:
    print(f"Type Error: {ae}")

except ValueError as ve:
    print(f"Invalid input: {ve}")

except Exception as e:
    print(f"Unexpected error: {e}")
```
